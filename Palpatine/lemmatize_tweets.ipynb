```python
import nltk
import pandas as pd
from nltk.tokenize import RegexpTokenizer
```

```python
df = pd.read_csv('cleaned_data.csv')
df = df[:10]
df
```

```python
def lemmatize_tweets(df):
    
    tokenizer = RegexpTokenizer(r'\w+')

    for index, row in df.iterrows():
        tweet = row["text"]
        tweet = tweet.lower()
        tweet = tokenizer.tokenize(tweet)
        text = ""
        for word in tweet:
            with open("odm.txt", "r", encoding="utf-8") as odm:
                for line in odm:
                    line = line.lower()
    #               line = tokenizer.tokenize(line)
    #               if word in line:
    #                     text += line[0] + " "
    #                     break
                    line = line.split(sep=", ")
                    line[len(line)-1] = line[len(line)-1].rstrip("\n")
                    if word in line:
                        text += line[0] + " "
                        break

        df.at[index, 'text'] = text
        
        return df
```

```python
df = lemmatize_tweets(df)
```

```python
df.to_csv('lemmatized_data.csv', index=False)
```
